```python
import glob
from pandas import DataFrame as df
import numpy as np
```

```python
items=glob.glob(r'..\PlantVillage-Dataset\raw\color\*\*.JPG',recursive=True)
print('Total no. of files='+str(len(items)))
```

```python
data=df(columns=['Item'],data=items)
data['Folder']=data['Item'].str.rsplit('\\', 1, expand=True)[0]
data['Folder']=data['Folder'].str.rsplit('\\', 1, expand=True)[1]
data['Species']=data['Folder'].str.rsplit('___', 1, expand=True)[0]
data['Disease']=data['Folder'].str.rsplit('___', 1, expand=True)[1]
data['Image']=data['Item'].str.rsplit('\\', 1, expand=True)[1]
data.drop(['Item','Folder'],axis=1,inplace=True)
```

```python
print('No. of species='+str(data.Species.nunique()))
```

```python
print('No. of diseases per species:\n'+str(data.groupby('Species')['Disease'].nunique()))
```
